```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
full_docs = docs[0].page_content + search_result + wiki_result
full_docs
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
docs = splitter.create_documents([full_docs])
```

```python
docs = splitter.split_documents(docs)
```

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ['HUGGING_FACE'] = os.getenv('HUGGING_FACE')
```

```python
from langchain_huggingface.embeddings import HuggingFaceEmbeddings

embedding = HuggingFaceEmbeddings(model_name='all-MiniLM-L6-v2')
```

```python
from langchain_community.vectorstores import FAISS

vectore_store = FAISS.from_documents(docs, embedding)
```

```python
vectore_store
```

```python
os.environ['GROQ_API_KEY'] = os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(model="llama3-groq-70b-8192-tool-use-preview")
```

```python
retriever = vectore_store.as_retriever()
```

```python
from langchain.prompts import PromptTemplate

prompt_temaplate = """ 
You are an expert AI assistant, Given the context, answer the question thoughtfully and try to give the answer briefly and with the topics name with bold.
there should be at least 5-6 topics and with the context there should be at least 10000 word in each topics.

Context: {context}

Question: {query}

Answer:
"""

prompt = PromptTemplate(
    input_variables=['context', "query"],
    template=prompt_temaplate
)
```

```python
from langchain.chains import RetrievalQA
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type='stuff',
    retriever=retriever,
    return_source_documents=True,
)
```

```python
query = "what is computer vision"

result = qa_chain({"query": query})
```
